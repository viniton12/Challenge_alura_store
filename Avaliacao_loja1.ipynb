```python
# importações de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loja1 = pd.read_csv('loja_1.csv')

df = pd.DataFrame(loja1)

display(loja1)
```

```python
#calcúlo do faturamento total de cada loja

def faturamento(df): #  uam função que calcula o faturamento da loja recebendo o paramêtro preço
    soma = df['Preço'].sum() # utilizamos sum() para calcúlar uma lista inteira do DataFrame
    return soma # retorna a soma da coluna preço 

fatura_total = faturamento(df) #C
print(f'o faturamento total da loja 1 foi de R$ {fatura_total}')
print('===' * 30)


# quantidade de vendas por categorias
#display(df['Categoria do Produto'].value_counts())

# ver a categoria mais vendida valor
contagem = df["Categoria do Produto"].value_counts()
print(contagem)
print('===' * 50)
categoria_mais_vendida_valor = contagem.max() # mostra o valor da categoria mais vendida
categoria_menos_vendida_valor= contagem.min() #mostra o valor da categoria menos vendida

# Categoria mais e menos vendida
categoria_mais_vendida = contagem.idxmax() #mostra o nome da categoria mais vendida
categoria_menos_vendida = contagem.idxmin() #mostra o nome da categoria menos vendida

print('===' * 50)
print(f' A categoria mais vendida foi {categoria_mais_vendida} com {categoria_mais_vendida_valor} unidades vendidas.')
print(f'A categoria menos vendida foi {categoria_menos_vendida} com {categoria_menos_vendida_valor} unidades vendidas.')

print('===' * 50)

# análise de avaliação do cliente
ava = display(df['Avaliação da compra'].value_counts())#exibe todas as avaliações dos clientes

media_avaliacao_loja = df['Avaliação da compra'].sum()/len(df['Avaliação da compra']) #tira á media de avaliação

print(f'A média de avaliação da loja foi {media_avaliacao_loja:.2f}')
print('===' * 50)

# em percentual
display(df['Avaliação da compra'].value_counts(normalize=True))  # exibi o percentual de avaliação da loja 
print('===' * 50)


#custo médio do frete para a loja 
custo_medio_frete =df['Frete'].sum()/len(df['Frete'])
print(f'O custo médio do frente  é de R$ {custo_medio_frete:.2f}.')
print('===' * 50)

```

```python
#  GRÁFICO  de número de vendas por categorias 

# plot
plt.figure(figsize=(50, 5))

#grafico de dispersão para categoria do produto e o número de vendas
plt.subplot(131)
plt.scatter(contagem.index, contagem.values, color='blue')
#difinindo legendas e plotando o gráfico
plt.xticks(rotation=30, fontsize=10) # Rotaing axis ticks and customizing their font size
for i, v in enumerate(contagem.values):
    plt.text(i, v + 5, str(v), ha='center', color='black') #Adicionar os valores no topo das barras
plt.title('Número de vendas por categoria', color= 'darkred', fontsize= 15)
plt.ylabel('Quantidade', color= 'red')
plt.xlabel('Categória', color= 'red')
plt.show()


#GRÁFICO DE BARRAS
plt.figure(figsize=(50, 5))
plt.subplot(132)
plt.bar(contagem.index, contagem.values, color= 'skyblue')
#difinindo legendas e plotando o gráfico
plt.xticks(rotation=30, fontsize=10) # Rotaing axis ticks and customizing their font size
for i, v in enumerate(contagem.values):
    plt.text(i, v + 5, str(v), ha='center', color='black') #Adicionar os valores no topo das barras
plt.title('Número de vendas por categoria', color= 'darkred', fontsize= 15)
plt.ylabel('Quantidade', color= 'red')
plt.xlabel('Categória', color= 'red')
plt.show()

#GRÁFICO DE LINHAS
plt.figure(figsize=(50, 5))
plt.subplot(133)
plt.plot(contagem.index, contagem.values, color= 'black')
#difinindo legendas e plotando o gráfico
plt.xticks(rotation=30, fontsize=10) # Rotaing axis ticks and customizing their font size
for i, v in enumerate(contagem.values):
    plt.text(i, v + 5, str(v), ha='center', color='black') #Adicionar os valores no topo das barras
plt.title('Número de vendas por categoria', color= 'darkred', fontsize= 15)
plt.ylabel('Quantidade', color= 'red')
plt.xlabel('Categória', color= 'red')
plt.show()
```

```python
#GRáfico de pizza para avaliação dos usuários 
# Contagem das avaliações
ava = df['Avaliação da compra'].value_counts()

# Criando o gráfico
fig, ax = plt.subplots(figsize=(7, 7))  # Tamanho ideal
explode = (0.1, 0, 0, 0, 0)  #separa a fatia e dá um destaque
ax.pie(ava.values, explode= explode,  labels=ava.index, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal') # para a pizza ficar redondinha
plt.title('Avaliação da loja',  color= 'darkred', fontsize= 15)
plt.show()


```

```python
#número de vendas por produtos
produtos = df['Produto'].value_counts()
display(produtos)

import pandas as pd

# Supondo que você tenha o DataFrame 'df' com os dados carregados
df['Faturamento'] = df['Preço'] * df['Quantidade de parcelas']  # ou 'Quantidade', dependendo do seu caso

# Agrupar por categoria e somar os faturamentos
faturamento_por_categoria = df.groupby('Categoria do Produto')['Faturamento'].sum().reset_index()

# Exibir o resultado
print(faturamento_por_categoria)

# Agrupar por produto e somar os faturamentos
faturamento_por_produto = df.groupby('Produto')['Faturamento'].sum().reset_index()

# Exibir o resultado
print(faturamento_por_produto)

import matplotlib.pyplot as plt



# Gráfico de faturamento por categoria
plt.figure(figsize=(15, 5))
plt.bar(faturamento_por_categoria['Categoria do Produto'], faturamento_por_categoria['Faturamento'], color='c')
plt.xlabel('Categoria')
plt.ylabel('Faturamento')
plt.title('Faturamento por Categoria')
plt.xticks(rotation=90)
plt.show()

# Gráfico de faturamento por produto
plt.figure(figsize=(15, 5))
plt.bar(faturamento_por_produto['Produto'], faturamento_por_produto['Faturamento'], color='g')
plt.xlabel('Produto')
plt.ylabel('Faturamento')
plt.title('Faturamento por Produto')
plt.xticks(rotation=90)
plt.show()

#GRÁFICO DE BARRAS vendas por produtos
plt.figure(figsize=(15, 5))

plt.bar(produtos.index, produtos.values, color= 'skyblue')
#difinindo legendas e plotando o gráfico
plt.xticks(rotation=90, fontsize=10) # Rotaing axis ticks and customizing their font size
for i, v in enumerate(produtos.values):
    plt.text(i, v + 5, str(v), ha='center', color='black') #Adicionar os valores no topo das barras
plt.title('Número de vendas por produto', color= 'darkred', fontsize= 15)
plt.ylabel('Quantidade', color= 'red')
plt.xlabel('Produto', color= 'red')
plt.show()


```
