O presente relatório tem como objetivo analisar o desempenho das quatro lojas avaliadas, com base em métricas relevantes como faturamento total, categorias de produtos mais e menos vendidos, média de avaliações dos clientes, frete médio e desempenho individual dos produtos. **Desenvolvimento**

1. Faturamento Total
A análise do faturamento total demonstrou que:

Loja_1 apresentou o maior faturamento.

Loja_4 teve o menor faturamento entre as quatro unidades.

Lojas 2 e 3 apresentaram valores muito próximos, ambos com bom desempenho.

2. Categorias de Produtos Mais e Menos Vendidas

Certas categorias específicas são recorrentes entre os produtos mais vendidos, como beleza, moda e acessórios.

Loja_4 teve a melhor avaliação média, demonstrando maior satisfação do cliente.

Loja_1 foi a pior avaliada, apesar do alto faturamento.

As Lojas 2 e 3 mantiveram médias semelhantes e consistentes.

4. Produtos Mais e Menos Vendidos
A análise individual revelou:

Algumas lojas têm um número alto de vendas concentradas em poucos produtos populares, o que pode indicar dependência.

Outras possuem uma distribuição mais equilibrada, o que reduz riscos.

5. Custo Médio do Frete
O custo médio do frete é um fator crucial:

Loja_1 apresentou o maior custo médio de frete, impactando diretamente o lucro líquido.

Loja_4 teve o menor custo de frete.

Lojas 2 e 3 se posicionaram no meio-termo, com custos otimizados.
