fazer gráficos de comparação:
1. faturamento das 4 lojas.
2. custo médio do frete
3. avaliação  édia das 4 lojas
4. ver quais lojas teve mais avaliação positiva e negativa.
5. ver os produtos mais vendido e menos vendido em cada loja. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Data Frame com as informações das lojas 
data = {
    'Loja':['loja_1', 'Loja_2', 'Loja_3', 'Loja_4'], 
    'Faturamento':[1534509.12, 1488459.06, 1464025.03, 1384497.58],
    'Frete médio':[34.69, 33.62,  33.07, 31.28],
    'Avaliação':[ 3.98, 4.04, 4.05, 4.00]
}

df = pd.DataFrame(data)
contagem = df['Loja'].value_counts() # criação do df

display(df) # exibição do Data frame

#GRÁFICO DE COMPARAÇÃO FATURAMENTO
plt.figure(figsize=(15, 5))
plt.bar(df['Loja'], df['Faturamento'], color='orange', width=0.3)
for i, v in enumerate(df['Faturamento']):
    plt.text(i, v + 3, str(v), ha='center', color='black')
plt.title('Comparação do faturamento das lojas', fontsize='15', color='black')
plt.xlabel('Lojas', fontsize=12, color='black')
plt.ylabel('Faturamento R$', fontsize=12, color='black')
plt.grid(True)
plt.show()

```

```python
#GRÁFICO DE AVALIAÇÃO DAS LOJAS
#GRÁFICO DE LINHAS
plt.figure(figsize=(8, 6))
plt.tight_layout(pad=2.0)  # Ajuste da margem interna
plt.bar(df['Loja'], df['Avaliação'], color= '#2C3E50', width=0.3)
#difinindo legendas e plotando o gráfico
plt.xticks(rotation=30, fontsize=10) # Rotaing axis ticks and customizing their font size

#legendas
plt.title('Comparação de avaliações das lojas', color= 'darkred', fontsize= 15)
plt.ylabel('Nota', color= 'red', fontsize= 15)
plt.xlabel('Loja', color= 'red', fontsize= 15)
plt.grid(True)
plt.show()


# GRÁFICO PARA AVALIAR O CUSTO DO FRETE DAS LOJAS
plt.figure(figsize=(8,6))
plt.bar(df['Loja'], df['Frete médio'], color='red', width=0.5)
plt.xticks(rotation=30, fontsize=10) # Rotaing axis ticks and customizing their font size

#legendas
plt.title('Custo Médio do frete', color= 'darkred', fontsize= 15)
plt.ylabel('Custo (R$)', color= 'red', fontsize= 15)
plt.xlabel('Loja', color= 'red', fontsize= 15)
plt.grid(True)
plt.show()
```

 Análise Comparativa das Lojas
 
Os gráficos de comparação visam demonstrar o desempenho das quatro lojas em relação ao faturamento, custo médio do frete, avaliação média, distribuição de avaliações positivas e negativas, e aos produtos mais e menos vendidos em cada loja.

A Loja_1 apresentou o maior faturamento, porém teve a menor avaliação média e o maior custo médio de frete.
Em contrapartida, a Loja_4 teve o menor custo médio de frete e o menor faturamento entre as quatro lojas, mas se destacou por ser a mais bem avaliada.

As Lojas 2 e 3 apresentaram faturamentos semelhantes — R$ 1.488.459,06 e R$ 1.464.025,03, respectivamente —, além de notas de avaliação próximas, indicando um desempenho equilibrado entre ambas.

Com o objetivo de reduzir custos operacionais, propõe-se o fechamento da Loja_1, que, embora tenha o maior faturamento, possui também o maior custo médio de frete, o que impacta negativamente o lucro líquido.

As Lojas 2, 3 e 4, além de apresentarem faturamento próximo, se destacam por terem melhores avaliações e menor custo médio de frete, o que contribui para aumentar a margem de lucro líquida dessas unidades.
